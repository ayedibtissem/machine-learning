```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential

img_rows, img_cols = 224, 224

# Définir les chemins des dossiers
train_dir = 'C:/Users/ayed_/Desktop/Train'
test_dir = 'C:/Users/ayed_/Desktop/Test'

# Définition du générateur d'images
train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2,  # 80% pour l'entraînement, 20% pour la validation
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_rows, img_cols),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_rows, img_cols),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_rows, img_cols),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)
```

```python

import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image dimensions
img_rows, img_cols = 224, 224

# Load the pre-trained DenseNet121 model without the top layers
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(img_rows, img_cols, 3))

# Add custom layers
model = Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(512, activation='relu'))  # Increased number of units
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))  # Additional dense layer
model.add(Dropout(0.5))
model.add(Dense(47, activation='softmax'))
# Compile the model
optimizer = Adam(learning_rate=0.0001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python

batch_size = 32
# Définir le nombre d'étapes par époque
train_steps_per_epoch = len(train_generator) // batch_size
val_steps_per_epoch = len(validation_generator) // batch_size
# Entraîner le modèle
history = model.fit(
    train_generator,
    steps_per_epoch=train_steps_per_epoch,
    epochs=30,  # Augmentation du nombre d'époques
    validation_data=validation_generator,
    validation_steps=val_steps_per_epoch
)
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Évolution de la perte')
plt.xlabel('Époque')
plt.ylabel('Perte')
plt.legend(['Entraînement', 'Validation'], loc='upper right')

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Évolution de exactitude')
plt.xlabel('Époque')
plt.ylabel('Exactitude')
plt.legend(['Entraînement', 'Validation'], loc='lower right')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
# Évaluer le modèle sur le jeu de test
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Afficher la matrice de confusion
conf_matrix = confusion_matrix(true_classes, predicted_classes)
print("Matrice de confusion :")
print(conf_matrix)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Faire des prédictions sur le jeu de test
y_pred = model.predict(test_generator)
y_true = test_generator.classes

# Convertir les prédictions en classes
y_pred_classes = np.argmax(y_pred, axis=1)

# Calculer les métriques de performance
accuracy = accuracy_score(y_true, y_pred_classes)
precision = precision_score(y_true, y_pred_classes, average='macro')
recall = recall_score(y_true, y_pred_classes, average='macro')
f1 = f1_score(y_true, y_pred_classes, average='macro')

print(f"Exactitude : {accuracy:.4f}")
print(f"Précision moyenne : {precision:.4f}")
print(f"Rappel moyen : {recall:.4f}")
print(f"Score F1 moyen : {f1:.4f}")

```
